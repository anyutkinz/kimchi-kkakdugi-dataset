# **Creating the Kimchi-Kkakdugi Dataset**

In this project, we will also explore how to create a small image dataset that you can use for your own educational projects. For my example, I chose to create a dataset of images of **kimchi** and **kkakdugi**.

> **What are Kimchi and Kkakdugi?**  
> **Kimchi** is a traditional Korean side dish, usually made from fermented napa cabbage and Korean radishes with a variety of seasonings.  
> **Kkakdugi** is a type of kimchi made specifically from cubed radish.  
> **The difference:** While both are types of kimchi, *kimchi* typically refers to the spicy fermented cabbage variety, whereas *kkakdugi* is made from diced radish and has a crunchier texture.

The result of this notebook is a small image dataset with **two classes**: Kimchi and Kkakdugi. Each class contains 1000 images.

Of course, you can feel free to create any dataset of your choice by following this tutorial. Using the provided scripts, you can also create datasets with more classes or larger sizes.

### **3. Image Augmentation:**

After scraping, we had about 300 images per class. Unfortunately, this is not enough even to train pretrained models, we are not even talking about "from scratch" models.  
To address this, we use **augmentation**.

> **What is augmentation?**  
> Augmentation is the process of artificially increasing the size and diversity of your dataset by applying random transformations such as rotations, flips, brightness/contrast changes, and zooms to existing images.  
> **Caution:** You shouldn't generate too many augmented images from a small original set, as this can lead to overfitting or lack of true diversity.  

See `augment_images.py` for the random augmentation logic used here.

```python
from augment_images import save_augmented_images
```

```python
folder_path = "kimchi"
save_augmented_images(folder_path, 1000)
```

```python
folder_path = "kkakdugi"
save_augmented_images(folder_path, 1000)
```

Let's take a look at few random images from our brand new dataset.

```python
import matplotlib.pyplot as plt
import random 
from PIL import Image

files = [f for f in os.listdir("kimchi") if f.endswith('.jpg')]
sample = random.sample(files, 5)

plt.figure(figsize=(15, 10))
for i, file in enumerate(sample):
    img = Image.open(os.path.join("kimchi", file))
    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.axis('off')
    
plt.show()
```

```python

files = [f for f in os.listdir("kkakdugi") if f.endswith('.jpg')]
sample = random.sample(files, 5)

plt.figure(figsize=(15, 10))
for i, file in enumerate(sample):
    img = Image.open(os.path.join("kkakdugi", file))
    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.axis('off')

plt.show()
```
